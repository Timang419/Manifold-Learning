```python
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap, TSNE
import umap # Make sure umap is installed: pip install umap-learn

# --- Configuration ---

# Define sample sizes to test
# You can adjust this list. Larger sizes will take longer to run.
# Start with a smaller range to test, then expand if needed.
# sample_sizes = [2000, 5000, 10000, 15000, 20000]
sample_sizes = [2000, 4000, 6000, 8000, 10000, 12000,15000, 20000] # Example range


# Use n_jobs=-1 to potentially speed up methods that support parallelization
# Set to 1 if you want purely single-core performance measurement
N_JOBS = -1 # Use all available CPU cores
# N_JOBS = 1 # Use only one CPU core

# Set a seed for reproducibility of the random subsampling across runs
RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

# --- Data Loading ---

# Label mapping (optional, not used in timing)
fashion_labels = {
    0: "T-shirt/top", 1: "Trouser", 2: "Pullover", 3: "Dress", 4: "Coat",
    5: "Sandal", 6: "Shirt", 7: "Sneaker", 8: "Bag", 9: "Ankle boot"
}

# 1. Load the FULL Fashion-MNIST dataset ONCE
print("Loading Fashion-MNIST dataset (this may take a moment)...")
# Using parser='liac-arff' might be necessary depending on your sklearn version and backend
try:
    fmnist = fetch_openml('Fashion-MNIST', version=1, as_frame=False, parser='liac-arff')
except Exception:
    print("Failed loading with liac-arff parser, trying default...")
    fmnist = fetch_openml('Fashion-MNIST', version=1, as_frame=False)

X_full = fmnist.data
y_full = fmnist.target.astype(int) # y not needed for timing, but good to have
print(f"Dataset loaded with {len(X_full)} samples.")
full_indices = np.arange(len(X_full))

# --- Scalability Testing ---

# Dictionary to store results: {method_name: [time_for_size1, time_for_size2, ...]}
results = {
    "PCA": [],
    "Isomap": [],
    "t-SNE": [],
    "UMAP": []
}

# Loop through each defined sample size
for n_samples in sample_sizes:
    print(f"\n--- Processing {n_samples} samples ---")

    # Subsample the data: Choose n_samples random indices WITHOUT replacement
    # Using the same indices for all methods within this loop ensures
    # they operate on the exact same data subset for fair comparison at this size.
    current_indices = np.random.choice(full_indices, n_samples, replace=False)
    X = X_full[current_indices]
    # y = y_full[current_indices] # Not needed for timing fit_transform

    # --- Time PCA ---
    print("Timing PCA...")
    start = time.time()
    pca = PCA(n_components=2, random_state=RANDOM_SEED)
    pca.fit_transform(X) # Perform the operation
    duration = time.time() - start
    results["PCA"].append(duration)
    print(f"PCA completed in: {duration:.4f} seconds")

    # --- Time Isomap ---
    # Isomap can be slow for larger N or complex manifolds
    print("Timing Isomap...")
    start = time.time()
    # Isomap benefits from n_jobs for neighbor search
    iso = Isomap(n_components=2, n_neighbors=15, n_jobs=N_JOBS)
    iso.fit_transform(X)
    duration = time.time() - start
    results["Isomap"].append(duration)
    print(f"Isomap completed in: {duration:.4f} seconds")

    # --- Time t-SNE ---
    # t-SNE (especially Barnes-Hut) complexity depends on N, perplexity
    print("Timing t-SNE...")
    start = time.time()
    # Barnes-Hut t-SNE (default method) benefits significantly from n_jobs
    tsne = TSNE(n_components=2, init="pca", random_state=RANDOM_SEED, perplexity=30, n_jobs=N_JOBS)
    tsne.fit_transform(X)
    duration = time.time() - start
    results["t-SNE"].append(duration)
    print(f"t-SNE completed in: {duration:.4f} seconds")

    # --- Time UMAP ---
    print("Timing UMAP...")
    start = time.time()
    # UMAP often benefits from n_jobs too
    umap_reducer = umap.UMAP(n_components=2, random_state=RANDOM_SEED, n_jobs=N_JOBS)
    umap_reducer.fit_transform(X)
    duration = time.time() - start
    results["UMAP"].append(duration)
    print(f"UMAP completed in: {duration:.4f} seconds")

print("\n--- Scalability Testing Complete ---")

# --- Plotting Results ---

plt.figure(figsize=(12, 8)) # Make plot slightly larger

# Plot time vs sample size for each method
for method_name, times in results.items():
    plt.plot(sample_sizes, times, marker='o', linestyle='-', linewidth=2, label=method_name)

# Add plot labels and title
plt.xlabel("Number of Samples", fontsize=12)
plt.ylabel("Execution Time (seconds)", fontsize=12)
plt.title("Scalability Comparison of Dimensionality Reduction Methods", fontsize=14)

# Add legend
plt.legend(fontsize=10)

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.7)

# Optional: Use a logarithmic scale for the y-axis if times vary widely
# plt.yscale('log')
# plt.ylabel("Execution Time (seconds, log scale)")

# Ensure x-axis ticks match the sample sizes tested
plt.xticks(sample_sizes, rotation=45)
plt.minorticks_off() # Turn off minor ticks if rotation overlaps them

plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()

print("\nExecution Time Summary (seconds):")
print(f"Sample Sizes: {sample_sizes}")
for method, times in results.items():
    formatted_times = [f"{t:.3f}" for t in times]
    print(f"{method}: {formatted_times}")
```
